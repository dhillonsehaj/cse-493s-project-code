```python
random_seed = 12
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

file_path = "dataset/emotional-reactions-reddit-model.csv"
data = pd.read_csv(file_path)

# 2. First, split the data into training (75%) and temp (25%) sets
train_data, temp_data = train_test_split(data, test_size=0.25, random_state=random_seed)

# 3. Now, split the temp_data into validation (5%) and test (20%) sets
val_data, test_data = train_test_split(temp_data, test_size=0.8, random_state=random_seed)

# 4. Save the three datasets to separate CSV files
train_data.to_csv("dataset/split/train_ER.csv", index=False)
val_data.to_csv("dataset/split/val_ER.csv", index=False)
test_data.to_csv("dataset/split/test_ER.csv", index=False)

file_path = "dataset/explorations-reddit-model.csv"
data = pd.read_csv(file_path)

# 2. First, split the data into training (75%) and temp (25%) sets
train_data, temp_data = train_test_split(data, test_size=0.25, random_state=random_seed)

# 3. Now, split the temp_data into validation (5%) and test (20%) sets
val_data, test_data = train_test_split(temp_data, test_size=0.8, random_state=random_seed)

# 4. Save the three datasets to separate CSV files
train_data.to_csv("dataset/split/train_EX.csv", index=False)
val_data.to_csv("dataset/split/val_EX.csv", index=False)
test_data.to_csv("dataset/split/test_EX.csv", index=False)

file_path = "dataset/interpretations-reddit-model.csv"
data = pd.read_csv(file_path)

# 2. First, split the data into training (75%) and temp (25%) sets
train_data, temp_data = train_test_split(data, test_size=0.25, random_state=random_seed)

# 3. Now, split the temp_data into validation (5%) and test (20%) sets
val_data, test_data = train_test_split(temp_data, test_size=0.8, random_state=random_seed)

# 4. Save the three datasets to separate CSV files
train_data.to_csv("dataset/split/train_IP.csv", index=False)
val_data.to_csv("dataset/split/val_IP.csv", index=False)
test_data.to_csv("dataset/split/test_IP.csv", index=False)
```
